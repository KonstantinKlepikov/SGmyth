```python
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
import os
```

```python
path = '../input/ruthenia'
names = os.listdir(os.path.realpath(path))
```

```python
def html_parser(names, path):
    partial = SoupStrainer('div', id='main')
    for name in names:
        npath = os.path.realpath(path) + '\\' + name
        with open(npath, 'r', encoding='utf-8') as file_open:
            try:
                html_obj = BeautifulSoup(file_open, 'html.parser', parse_only=partial)
            except AttributeError:
                print('not a html')         
            print(html_obj.prettify(), '\n')

            motiv = html_obj.find('p', {'class': 'NormalLin'}).get_text()
            description = html_obj.find('p', {'class': 'NormalLis'}).get_text()
            stories = html_obj.find_all('p', {'class': 'NormalMai'})
            print(motiv)
            print(description)
            for story in stories:
                if story.b:
                    region = story.b
                    if isinstance(region.next_element.next_element, str):
                        region_next = region.next_element
                        region_text1 = story.b.get_text() 
                        region_text2 = region.next_element.next_element
                    else:
                        region_next = region
                        region_text1 = story.b.get_text()
                        region_text2 = None 
                else:
                    region = None
                peoples = region_next.find_next('u')
                peoples_text = region_next.next_element.next_element.next_element.next_element
                peoples = peoples.get_text()
                
#                 for people in peoples:
#                     people_next = people.get_text()
#                     print(people, people.find_next())
#                     if isinstance(people.find_next(), str):
#                         text_next = people.find_next()
#                     else:
#                         text_next = None
#                     print(people_next, text_next)
#                 substory = peoples[0].find_all('i')
                print('region_text1:' + region_text1)
                print('region_text2:' + region_text2)
                print('peoples:' + peoples)
                print('peoples_text:' + peoples_text)
    
    
            yield print('ok')
```

```python
parsed = html_parser(names, path)
```
