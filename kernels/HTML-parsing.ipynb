```python
from bs4 import BeautifulSoup
import os
```

```python
path = '../input/ruthenia'
names = os.listdir(os.path.realpath(path))
```

```python
def html_parser(names, path):
    for name in names:
        npath = os.path.realpath(path) + '\\' + name
        with open(npath, 'r', encoding='utf-8') as file_open:
            try:
                html_obj = BeautifulSoup(file_open, 'html.parser')
            except AttributeError:
                print('not a html')         
            print(html_obj.body, '\n')

            inside = html_obj.find_all('div', {'id': 'main'})

            NormalLin = html_obj.find_all('p', {'class': 'NormalLin'})
            NormalLis = html_obj.find_all('p', {'class': 'NormalLis'})
            NormalMai = html_obj.find_all('p', {'class': 'NormalMai'})

            yield print(NormalLin, '\n', NormalLis, '\n', NormalMai, '\n')
```

```python
parsed = html_parser(names, path)
```
